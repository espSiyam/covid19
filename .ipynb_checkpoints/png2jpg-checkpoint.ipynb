```python
import os 
from PIL import Image
from tqdm import tqdm

images = os.listdir('./images')
length = len(images)
length
```

```python
Image.open('./images/'+images[0])
```

```python
name = os.path.splitext(images[0])[0]
```

```python
im = Image.open('./images/'+images[0])
rgb_im = im.convert('RGB')
name = os.path.splitext(images[0])[0]
rgb_im.save('./convert/'+name+'.jpg')
```

```python
for i in tqdm(range(length)):
    im = Image.open('./images/'+images[i])
    rgb_im = im.convert('RGB')
    name = os.path.splitext(images[i])[0]
    rgb_im.save('./convert/'+name+'.jpg')
```
